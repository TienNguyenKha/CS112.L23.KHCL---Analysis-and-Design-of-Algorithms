## Algorithm Efficiency
- Hàm `sorted()` được build sẵn trong `Python` có độ phức tạp là:
  - Best-case: `O(n)`
  - Average-case: `O(nlogn)`
  - Worst-case: `O(nlogn)`

  Vì vậy nếu `n` (độ dài của mảng `a`) đủ lớn thì sẽ rơi vào một trong hai trường hợp dưới.
- Hàm `H_index()` có độ phức tạp là `O(n)`
- Cả chương trình sẽ có độ phức tạp là: `O(nlogn + n)`, tùy thuộc vào độ dài của mảng `a`.

## Programming

**Input:** 
- Dòng đầu tiên chứa một số nguyên ***n*** (1 ≤ ***n*** ≤ 5×10<sup>5</sup>)),
- Dòng thứ 2 chứa ***n*** số nguyên ***c<sub>1</sub>, c<sub>2</sub>, . . ., c<sub>n</sub>*** (0 ≤ ***c<sub>i</sub>*** ≤ 10<sup>6</sup>, ***i*** = 1 ÷ ***n***).

```python
n = int(input())
a = sorted(list(map(int, input().split())))
```

Định nghĩa hàm `H_index`:

```python
def H_index(a, n):
    for i in range(n):
        result = n - i
        if result <= a[i]:
            return result
    return 0
```

**Output:**

```python
print(H_index(a, n))
```

## Algorithm Efficiency
- Hm `sorted()` c build sn trong `Python` c  phc tp l:
  - Best-case: `O(n)`
  - Average-case: `O(nlogn)`
  - Worst-case: `O(nlogn)`

  V vy nu `n` ( di ca mng `a`)  ln th s ri vo mt trong hai trng hp di.
- Hm `H_index()` c  phc tp l `O(n)`
- C chng trnh s c  phc tp l: `O(nlogn + n)`, ty thuc vo  di ca mng `a`.

## Programming

**Input:** 
- Dng u tin cha mt s nguyn ***n*** (1  ***n***  510<sup>5</sup>)),
- Dng th 2 cha ***n*** s nguyn ***c<sub>1</sub>, c<sub>2</sub>, . . ., c<sub>n</sub>*** (0  ***c<sub>i</sub>***  10<sup>6</sup>, ***i*** = 1  ***n***).

```python
n = int(input())
a = sorted(list(map(int, input().split())))
```

nh ngha hm `H_index`:

```python
def H_index(a, n):
    for i in range(n):
        result = n - i
        if result <= a[i]:
            return result
    return 0
```

**Output:**

```python
print(H_index(a, n))
```

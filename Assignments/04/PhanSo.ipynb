# Pattern Recognition
* Algorithm : Brute Force
* Signature : Bài toán yêu cầu biến đổi từng bước nên Brute Force là thuật toán tối ưu nhất để không phải bỏ qua bước nào, ta chỉ cần thêm điều kiệu để nó không bị chạy vô hạn lần

# Algorithm Design
* Đối với bài toán phân số này chúng ta phải có điều kiện dừng để Brute Force không bị chạy đến vô hạn

* Ta dễ dàng nhận thấy là khi phân số thứ nhất lớn hơn phân số thứ 2 thì không có cách nào tìm được các phép biến đổi phù hợp

* Khi phân số thứ 1 lớn hơn phân số thứ 2 thì ta cũng không thể tìm được các phép biến đổi phù hợp 
* Sau đó áp dụng các bước Brute Force cơ bản để tìm được số phép biến đổi phù hợp

# Algorithm Efficiency
* Best case(Phân số thứ 1 lớn hơn phân số thứ 2) : O(1) 
* Avarage case and worse case :  

# Code

```python
import math

a = int(input())
b = int(input())
c = int(input())
d = int(input())
```

```python
if (d*a - b*c) >= 0:
    print(0)
else:
    count = 0
    while ((a*d != b*c) and ((a*d - b*c) < 0)):
        a = a + 1
        b = b + 1
        gcd = math.gcd(a, b)
        a = a//gcd
        b = b//gcd
        count = count + 1
    if (a*d - b*c) > 0 :
        print(0)
    else:
        print(count)
```

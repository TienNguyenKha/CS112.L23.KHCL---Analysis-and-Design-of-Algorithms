## Abstraction
Tìm số Fibonacci thứ *n<sub>th</sub>*.

## Pattern Recognition


- **Tên dạng bài toán:** Tìm số Fibonacci thứ *n<sub>th</sub>*.

- **Đặc điểm nhận dạng:**

|Ngày|0|1|2|3|...|
|:-|:-:|:-:|:-:|:-:|:-:|
|Tảo|1|2|5|13|...|
| |1<sub>1</sub>|1<sub>1</sub>1<sub>2</sub>|2<sub>2</sub>3<sub>1</sub>|2<sub>3</sub>3<sub>2</sub>...|...|
|Fibonacci|1(1)|2(3)|5(5)|13(7)|...|

Sau khi vẽ bảng trên ta nhận ra rằng số lượng tảo trong ngày thứ *k* bằng với số Fibonacci *2k + 1*

## Algorithm Designed
 Performed by Pseudocode
 
``` python
def fibonacci:
  if n <= 1:
    return n
  a = 0
  b = 1
  c = 0

  for i in range(2, n + 1):
    c = a + b
    a = b
    b = c
```

## Algorithm Efficiency
- Với *n <= 1* thì độ phức tạp là *O(1)*.
- Với *n >= 2* thì thời gian thực hiện vòng `for` là *T(n - 2)*, suy ra độ phức tạp của giải thuật trên là *O(n)*



***Time Complexity:*** O(n)

***Extra Space:*** O(1)

## Programming

**Input:** *n* và *k* (1 ≤ n ≤ 1000, 1 ≤ k ≤ 10<sup>17</sup>)

````python
n, k = map(int, input().split())
````

Define `fibonacci` function with algorithm that has the **space opimized**.

````python
def fibonacci(n):
    if n <= 1:
        return n
    a = 0
    b = 1
    c = 0

    for i in range(2, n + 1):
        c = a + b
        a = b
        b = c

    return c
````

**Output:** 

````python
print(n * fibonacci(2 * k + 1) % (pow(10, 9) + 7))
````
